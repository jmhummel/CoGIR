```python
import albumentations as A
import matplotlib.pyplot as plt
import numpy as np

from pathlib import Path

from data.dataset import ImageDataset
```

```python
degradations = [
    A.RandomBrightnessContrast(p=1.0, brightness_limit=0.4, contrast_limit=0.4),
]
```

```python
# Load dataset
dataset = ImageDataset(location=Path('../tests/data/'), degradations=degradations)
```

```python
# Get an item from the dataset (first image)
input_tensor, target_tensor = dataset[0]
```

```python
# Convert tensors back to images for display
input_image = np.transpose(input_tensor.numpy(), (1, 2, 0))  # (H, W, C)
target_image = np.transpose(target_tensor.numpy(), (1, 2, 0))  # (H, W, C)
```

```python
# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title('Input Image')

plt.subplot(1, 2, 2)
plt.imshow(target_image)
plt.title('Target Image')

plt.show()
```
